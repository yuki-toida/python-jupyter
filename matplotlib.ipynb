```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(-3.41, 3.14, 0.25)
y1 = np.sin(x1)
x2 = np.linspace(0, 3, 10)
y2 = x2 + 1

# 描画領域の確保(横幅:800px、高さ:600px)
fig = plt.figure(figsize=(8.0, 6.0))
# 線プロット
plt.plot(x1, y1, "r-", lw=2, alpha=0.7, ms=2, label="(x1, y1)")
# 点プロット
plt.plot(x2, y2, "bo", lw=2, alpha=0.7, ms=5, label="(x2, y2)")
# 全体のフォント
plt.rcParams['font.family'] = 'Times New Roman'
# フォントサイズ
plt.rcParams['font.size'] = 20
# 軸の太さ
plt.rcParams['axes.linewidth'] = 1.0
# 凡例の表示（2：位置は第二象限）
plt.legend(loc=2, fontsize=20)
# グラフタイトル
plt.title('Graph Title', fontsize=20)
# x軸ラベル
plt.xlabel('x', fontsize=20)
# y軸ラベル
plt.ylabel('y', fontsize=20)
# x軸範囲
plt.xlim([-3, 3])
# y軸範囲
plt.ylim([-2, 4])
# 軸ラベルの目盛りサイズ
plt.tick_params(labelsize=20)
# x軸の目盛りを引く場所を指定（無ければ自動で決まる）
plt.xticks(np.arange(-3.0, 4.0, 1.0))
# y軸の目盛りを引く場所を指定（無ければ自動で決まる）
plt.yticks(np.arange(-3.0, 4.0, 1.0))
# x, y軸のスケールを均等
plt.axis('scaled')
# ラベルがきれいに収まるよう表示
plt.tight_layout()
# グリッドの表示
plt.grid()
# グラフ表示
plt.show()

# グラフをファイル保存
fig.savefig("dist/graph.png")
```
